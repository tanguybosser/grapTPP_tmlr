## Simulations

```python
import pickle 
import numpy as np 
import matplotlib.pyplot as plt 

def deviation_per_mark(model, dir, splits):
    model_types = ['', '-dd']
    marks = ['K1', 'K2', 'K3', 'K4', 'K5']
    all_dev, all_std = [], []
    for type in model_types:
        all_dev_per_mark = []
        for split in range(splits): 
            if type == '-dd':
                file = f'{dir}/hawkes_exponential_mutual_bis_gru_temporal_with_labels_gru_temporal_with_labels_{model}{type}_split{split}.txt'
            else:
                file = f'{dir}/hawkes_exponential_mutual_bis_gru_{model}{type}_temporal_with_labels_split{split}.txt'
            with open(file, 'rb') as f:
                results = pickle.load(f)
            all_dev_per_mark.append(results['marked'])
        all_dev_per_mark = np.concatenate([all_dev_per_mark], axis=0)
        mean_dev_per_mark = np.mean(all_dev_per_mark, axis=0)
        all_dev.append(mean_dev_per_mark)
        std_dev_per_mark = np.std(all_dev_per_mark, axis=0)
        all_std.append(std_dev_per_mark)
    all_dev = np.array(all_dev)
    all_std = np.array(all_std) 
    #all_dev = np.transpose(np.array(all_dev))
    #all_std = np.transpose(np.array(all_std))
    print(all_dev.shape)
    print(all_std.shape)
    n = all_dev.shape[0]
    r = np.arange(len(marks))
    width = 0.8
    for i in range(n):
        plt.bar(r - width/2. + i/float(n)*width, all_dev[i], 
                width=width/float(n), align="edge")   
    plt.xticks(r, marks)
    plt.show()
```

```python
model = 'thp'
dir = 'results/simulations2'
splits = 0

deviation_per_mark(model, dir, splits)
```
